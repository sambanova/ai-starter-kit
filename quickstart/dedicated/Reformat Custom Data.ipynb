```python
from IPython.display import display, HTML
display(HTML("<style>:root { --jp-notebook-max-width: 100% !important; }</style>"))
import json
import os
import pprint
import jsonlines
import pandas as pd
pd.set_option('display.max_columns', None)
```

## 2. Set source & destination paths

```python
source_path = os.path.expanduser("~/Downloads/single_prompt_data.json")
destination_path = os.path.expanduser("~/Downloads/custom_prompt_data.jsonl")
```

## 3. Reformat file

```python
with open(source_path, "r") as f:
    original_file_contents = json.load(f)
original_file_contents[0]
```

```python
output_prompts = []
for item in original_file_contents:
    output_prompts.append({"prompt": item['messages'][0]['content']})
len(original_file_contents), len(output_prompts)
```

```python
output_prompts["0"]
```

## 4. Write reformatted file to destination path

```python
with jsonlines.open(destination_path, "w") as writer:
       for item in output_prompts:
           writer.write(item)
```
